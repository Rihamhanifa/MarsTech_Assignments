```python
#import Libraries
import pandas as pd # Import required library for data manipulation
from sklearn.model_selection import train_test_split # Import library for splitting data into training and testing sets
from sklearn.linear_model import LinearRegression # Import library for linear regression model
from sklearn.metrics import r2_score # Import library for model evaluation
```

```python
# Load the student performance dataset from CSV file
dataset = pd.read_csv ("Student_Performance.csv") 
dataset
```

```python
 # One-hot encode categorical 'Extracurricular Activities' column
dataset = pd.get_dummies (dataset)
dataset
```

```python
dataset.columns
```

```python
#Splitting Data into Independent (X) and Dependent (Y) Variables
#split input 
independent = dataset [['Hours Studied', 'Previous Scores', 'Sleep Hours',
       'Sample Question Papers Practiced','Extracurricular Activities_Yes']] 
```

```python
 #Split output 
dependent = dataset[['Performance Index']]
```

```python
#Splitting Data into Training & Testing Sets
x_train,x_test,y_train,y_test = train_test_split(independent , dependent ,test_size = 0.30, random_state=0) 
```

```python
# Trains the model on training data
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
# Extract model coefficients (weights) for each feature
weight = regressor.coef_
weight
```

```python
# Extract model intercept (bias) for the linear equation
bias = regressor.intercept_
bias
```

```python
# Generate predictions on the test set
y_pred = regressor.predict(x_test)
y_pred
```

```python
# Calculate RÂ² score to evaluate model performance (higher is better)
r_score = r2_score (y_test,y_pred)
r_score
```
